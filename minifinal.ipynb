```python
# Import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")

# Drop rows with missing values
df = df.dropna()

# Simple Linear Regression using 'year' as the feature
X = df[['year']]  # Independent variable
y = df['selling_price']  # Dependent variable

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate
print("R Score:", r2_score(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))

# Visualize
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, y_pred, color='red', label='Predicted')
plt.title('Simple Linear Regression: Year vs Selling Price')
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error
import tkinter as tk
from tkinter import ttk, messagebox

class CarPricePredictor:
    def __init__(self, root):
        self.root = root
        self.root.title("Car Sales Predictor")
        self.root.geometry("600x500")
        
        # Load and preprocess data
        self.load_data()
        self.preprocess_data()
        self.train_model()
        
        # Create GUI
        self.create_widgets()
    
    def load_data(self):
        # Load the dataset (in a real app, you'd load from the CSV file)
        # For this example, we'll assume the data is loaded into self.df
        # In practice: self.df = pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
        pass
    
    def preprocess_data(self):
        # Sample preprocessing (actual implementation would use your CSV data)
        # Create a sample dataframe for demonstration
        data = {
            'name': ['Maruti 800 AC', 'Maruti Wagon R LXI Minor', 'Hyundai Verna 1.6 SX'],
            'year': [2007, 2007, 2012],
            'selling_price': [60000, 135000, 600000],
            'km_driven': [70000, 50000, 100000],
            'fuel': ['Petrol', 'Petrol', 'Diesel'],
            'seller_type': ['Individual', 'Individual', 'Individual'],
            'transmission': ['Manual', 'Manual', 'Manual'],
            'owner': ['First Owner', 'First Owner', 'First Owner']
        }
        self.df = pd.DataFrame(data)
        
        # Feature engineering
        self.df['age'] = 2023 - self.df['year']  # Assuming current year is 2023
        
        # Encode categorical variables
        self.encoders = {}
        categorical_cols = ['fuel', 'seller_type', 'transmission', 'owner']
        
        for col in categorical_cols:
            le = LabelEncoder()
            self.df[col] = le.fit_transform(self.df[col])
            self.encoders[col] = le
        
        # Select features and target
        self.features = ['age', 'km_driven', 'fuel', 'seller_type', 'transmission', 'owner']
        self.X = self.df[self.features]
        self.y = self.df['selling_price']
    
    def train_model(self):
        # Split data
        X_train, X_test, y_train, y_test = train_test_split(
            self.X, self.y, test_size=0.2, random_state=42
        )
        
        # Train model
        self.model = RandomForestRegressor(n_estimators=100, random_state=42)
        self.model.fit(X_train, y_train)
        
        # Evaluate
        y_pred = self.model.predict(X_test)
        mae = mean_absolute_error(y_test, y_pred)
        print(f"Model MAE: {mae:.2f}")
    
    def create_widgets(self):
        # Title
        title_label = ttk.Label(self.root, text="Car Price Predictor", font=("Arial", 16, "bold"))
        title_label.pack(pady=10)
        
        # Input Frame
        input_frame = ttk.LabelFrame(self.root, text="Car Details", padding=10)
        input_frame.pack(pady=10, padx=10, fill="both", expand=True)
        
        # Car Brand/Model
        ttk.Label(input_frame, text="Car Brand/Model:").grid(row=0, column=0, padx=5, pady=5, sticky="e")
        self.car_model = ttk.Entry(input_frame)
        self.car_model.grid(row=0, column=1, padx=5, pady=5, sticky="w")
        
        # Year
        ttk.Label(input_frame, text="Year:").grid(row=1, column=0, padx=5, pady=5, sticky="e")
        self.year = ttk.Entry(input_frame)
        self.year.grid(row=1, column=1, padx=5, pady=5, sticky="w")
        
        # Kilometers Driven
        ttk.Label(input_frame, text="Kilometers Driven:").grid(row=2, column=0, padx=5, pady=5, sticky="e")
        self.km_driven = ttk.Entry(input_frame)
        self.km_driven.grid(row=2, column=1, padx=5, pady=5, sticky="w")
        
        # Fuel Type
        ttk.Label(input_frame, text="Fuel Type:").grid(row=3, column=0, padx=5, pady=5, sticky="e")
        self.fuel_type = ttk.Combobox(input_frame, values=["Petrol", "Diesel", "CNG", "LPG", "Electric"])
        self.fuel_type.grid(row=3, column=1, padx=5, pady=5, sticky="w")
        
        # Seller Type
        ttk.Label(input_frame, text="Seller Type:").grid(row=4, column=0, padx=5, pady=5, sticky="e")
        self.seller_type = ttk.Combobox(input_frame, values=["Individual", "Dealer", "Trustmark Dealer"])
        self.seller_type.grid(row=4, column=1, padx=5, pady=5, sticky="w")
        
        # Transmission
        ttk.Label(input_frame, text="Transmission:").grid(row=5, column=0, padx=5, pady=5, sticky="e")
        self.transmission = ttk.Combobox(input_frame, values=["Manual", "Automatic"])
        self.transmission.grid(row=5, column=1, padx=5, pady=5, sticky="w")
        
        # Owner
        ttk.Label(input_frame, text="Owner:").grid(row=6, column=0, padx=5, pady=5, sticky="e")
        self.owner = ttk.Combobox(input_frame, values=[
            "First Owner", "Second Owner", "Third Owner", "Fourth & Above Owner"
        ])
        self.owner.grid(row=6, column=1, padx=5, pady=5, sticky="w")
        
        # Predict Button
        predict_btn = ttk.Button(self.root, text="Predict Price", command=self.predict_price)
        predict_btn.pack(pady=10)
        
        # Result Label
        self.result_label = ttk.Label(self.root, text="Predicted Price: ", font=("Arial", 12))
        self.result_label.pack(pady=10)
    
    def predict_price(self):
        try:
            # Get input values
            year = int(self.year.get())
            km_driven = int(self.km_driven.get())
            fuel = self.fuel_type.get()
            seller = self.seller_type.get()
            transmission = self.transmission.get()
            owner = self.owner.get()
            
            # Calculate age
            age = 2023 - year  # Update with current year
            
            # Encode categorical features
            fuel_encoded = self.encoders['fuel'].transform([fuel])[0]
            seller_encoded = self.encoders['seller_type'].transform([seller])[0]
            transmission_encoded = self.encoders['transmission'].transform([transmission])[0]
            owner_encoded = self.encoders['owner'].transform([owner])[0]
            
            # Create feature array
            features = np.array([[age, km_driven, fuel_encoded, seller_encoded, 
                                transmission_encoded, owner_encoded]])
            
            # Make prediction
            prediction = self.model.predict(features)
            
            # Display result
            self.result_label.config(text=f"Predicted Price: {prediction[0]:,.2f}")
            
        except ValueError as e:
            messagebox.showerror("Input Error", "Please enter valid numbers for year and kilometers driven")
        except Exception as e:
            messagebox.showerror("Error", f"An error occurred: {str(e)}")

if __name__ == "__main__":
    root = tk.Tk()
    app = CarPricePredictor(root)
    root.mainloop()
```
